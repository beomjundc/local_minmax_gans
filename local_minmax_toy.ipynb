```python
# Import libs
import numpy as np
import matplotlib.pyplot as plt
from autograd import grad

# Increase standard plot size and face color
plt.rcParams['figure.figsize'] = (12, 8)

# Define methods for computing actual value vs. theoretical bounds
def gradient_descent_ascent(max_iters, lr, decay, k, f, df_x, df_y, x_0, y_0, inf):
    f_k = []
    curr_x_k = []
    curr_y_k = []
    
    curr_x = x_0 # set current x to x_0
    curr_y = y_0

    while k < max_iters:
        prev_x = curr_x
        prev_y = curr_y
        
        curr_x = curr_x - (lr/(1 + decay))*df_x(prev_x, prev_y) # general gradient descent
        curr_y = curr_y + lr*df_y(prev_x, prev_y) # general gradient ascent
        
        print(curr_x)
        print(curr_y)
        
        f_k.append(f(curr_x, curr_y))
        curr_x_k.append(curr_x)
        curr_y_k.append(curr_y)
        
        k = k + 1
        
        if inf == True:
            decay = decay + 0.001
        else:
            decay = decay
    
    return f_k, curr_x_k, curr_y_k

def plot_loss_acc(result):
    f, ax1 = plt.subplots()
    ax1.plot(result[1])
    ax1.plot(result[2])

    ax1.legend(['x', 'y'], loc=0) # use best loc automatically
    ax1.set_title('Variables and O.F.V')
    ax1.set_xlabel('k')
    ax1.set_ylabel('x / y') 
    
    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    ax2.plot(result[0], c='g')
    ax2.set_ylabel('O.F.V')
    ax2.yaxis.label.set_color('g')
    ax2.tick_params(axis='y', colors='g')
    plt.plot()

# Define functions

# Toy Problem 1
f1 = lambda x, y: x**2 - y**2
df1_x = lambda x, y: 2*x
df1_y = lambda x, y: -2*y
ddf1_xx = 2
ddf1_yy = -2
ddf1_xy = 0
ddf1_yx = 0
# critical point is Nash equilibrium and local minmax

# Toy Problem 2
f2 = lambda x, y: -1*x**2 + 5*x*y - y**2
df2_x = lambda x, y: -2*x + 5*y
df2_y = lambda x, y: 5*x - 2*y
ddf2_xx = -2
ddf2_yy = -2
ddf2_xy = 5
ddf2_yx = 5
# critical point is local minmax
```

```python
# Set params
max_iters = 100 # max number of iterations
x_0 = 1
y_0 = 1

min_max1 = gradient_descent_ascent(max_iters, 0.1, 1, 0, f1, df1_x, df1_y, x_0, y_0, False)
min_max1
plot_loss_acc(min_max1)
```

```python
# Set params
max_iters = 100 # max number of iterations
x_0 = 1
y_0 = 1

min_max2 = gradient_descent_ascent(max_iters, 0.1, 1, 0, f1, df1_x, df1_y, x_0, y_0, True)
min_max2
plot_loss_acc(min_max2)
```
